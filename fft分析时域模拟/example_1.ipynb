```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = 'fangsong'
plt.rcParams['axes.unicode_minus'] = False
```

```python
Adc = 2                 # 直流分量幅度
A1 = 3                  # 频率F1信号的幅度
A2 = 1.5                # 频率F2信号的幅度
F1 = 50                 # 信号1频率(Hz)
F2 = 75                 # 信号2频率(Hz)
Fs = 512                # 采样频率(Hz)
P1 = -30                # 信号1相位(度)
P2 = 90                 # 信号相位(度)
N = 256*256             # 采样点数
Ts = 1/Fs               
t = np.arange(1, N*Ts, Ts)
```

```python
plt.figure(figsize=(5, 3), dpi=100)
S = Adc + A1*np.cos(2*np.pi*F1*t+np.pi*P1/180) \
        + A2*np.cos(2*np.pi*F2*t+np.pi*P2/180)

plt.plot(t[:200], S[:200])
plt.title('原始信号')
plt.show()
```

```python
plt.figure(figsize=(5, 3), dpi=100)
Y = scipy.fft.fft(S, N)             # Y为一个复数
Ayy = np.abs(Y)                     # 幅度
plt.plot(Ayy)
plt.show()
```

```python
plt.figure(figsize=(5, 3), dpi=100)
Ayy = Ayy/(N/2)                     # 换算成实际的幅度
Ayy[0]/=2  
F = (np.arange(N))*Fs/N             # 换算成实际的频率值
plt.plot(F[:N//2], Ayy[:N//2])        # 显示换算后的FFT模值结果
plt.title('幅度-频率曲线图')
plt.show()
```
